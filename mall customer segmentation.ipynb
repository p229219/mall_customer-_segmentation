```python
#clustering relationship between age and spending score
x1=df.loc[:,["Age","Spending Score (1-100)"]].values
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(x1)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(15,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth=4,color="blue",marker="8")
plt.xlabel("k value")
plt.ylabel("wcss")
plt.show()
```

```python
# initializing the class object
kmeans = KMeans(n_clusters=4)
#predict the labels for cluster
label=kmeans.fit_predict(x1)
print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(x1[:,0],x1[:,1],c=kmeans.labels_,cmap="rainbow")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color="brown")
plt.title("cluster of customer")
plt.xlabel("age")
plt.ylabel("spending")
plt.show()
```
